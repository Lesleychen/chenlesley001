```python
from AWSIoTPythonSDK.MQTTLib import AWSIoTMQTTShadowClient
import random, time, datetime
import pandas as pd
import numpy as np
import json
import os

# A random programmatic shadow client ID.
SHADOW_CLIENT = "A0206691A"

# The unique hostname that &IoT; generated for this device.
HOST_NAME = "albx8ncfasetk-ats.iot.ap-southeast-1.amazonaws.com"

# The relative path to the correct root CA file for &IoT;, 
# which you have already saved onto this device.
ROOT_CA = "Amazon_Root_CA_1.pem.txt"

# The relative path to your private key file that 
# &IoT; generated for this device, which you 
# have already saved onto this device.
PRIVATE_KEY = "9cad72e2b4-private.pem.key"

# The relative path to your certificate file that 
# &IoT; generated for this device, which you 
# have already saved onto this device.
CERT_FILE = "9cad72e2b4-certificate.pem.crt"

# A programmatic shadow handler name prefix.
SHADOW_HANDLER = "A0206691A"

# Automatically called whenever the shadow is updated.
def myShadowUpdateCallback(payload, responseStatus, token):
  print()
  print('UPDATE: $aws/things/' + SHADOW_HANDLER + 
    '/shadow/update/#')
  print("payload = " + payload)
  print("responseStatus = " + responseStatus)
  print("token = " + token)

# Create, configure, and connect a shadow client.
print('Connecting to AWS IoT!')

myShadowClient = AWSIoTMQTTShadowClient(SHADOW_CLIENT)
myShadowClient.configureEndpoint(HOST_NAME, 8883)
myShadowClient.configureCredentials(ROOT_CA, PRIVATE_KEY,
  CERT_FILE)
myShadowClient.configureConnectDisconnectTimeout(10)
myShadowClient.configureMQTTOperationTimeout(5)
myShadowClient.connect()
#MQTTClient=myShadowClient.getMQTTConnection()
#MQTTClient.configureOfflinePublishQueueing(500, yourDropBehavior)

# Create a programmatic representation of the shadow.
myDeviceShadow = myShadowClient.createShadowHandlerWithName(
  SHADOW_HANDLER, True)

# Keep generating random test data until this script stops running.

print('program start!')

engfile=open(r'e:\1st nus\ee5111\FD001.txt') #read the FDOO1.txt file
line=engfile.readlines()                     #read the txt file line by line
i = 0
while i < len(line):
    x = line[i].split()  #transform every line into a list   
    x[0] = 'FD001_'+x[0] #Overwrite column 'id' of the engine as 'FD001' + id (e.g. FD001_12) 
    y = {"id":x[0],"circle":x[1], "os1":x[2], "os2":x[3], "os3":x[4]}  #y is a dict with columns id, circle, os1,os2,os3
    for j in range(21):
        y["sensor"+str(j+1)]=x[j+5]                                    #y is a dict with columns sensor1-21
    t=time.time()
    t=time.strftime("%Y/%m/%d %H:%M:%S",time.localtime(t))
    y["timestamp"]= (t)                                               #add a column timestamp
    y["matric no."]= "A0206691A"   #add a column matric number
    y = {"state":{"reported":y}}
    print(y)
    j=json.dumps(y)                                                  #tranform y into a string in the form of json
    myDeviceShadow.shadowUpdate(j,myShadowUpdateCallback,5)     #update data onto thing shadow
    time.sleep(1)                                                    #update rate is set to be 1 sec
    i += 1
print("end")

engfile.close()
```
